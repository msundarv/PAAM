```python
# Read assets from a YAML file

with open("../data/assets.yml", "r") as file:
    assets = yaml.safe_load(file)

print(assets)
```

```python
# Test Price Pulse Search Tool

from tools import load_search_tool

search_tool = load_search_tool()

asset_key = "asset_001"
query = (
    "What is the current price of "
    + assets[asset_key]["asset"]
    + " per "
    + assets[asset_key]["investment_unit"]
    + " in "
    + assets[asset_key]["currency"]
    + "?"
)

response = search_tool.run(query)

print(response)
```
